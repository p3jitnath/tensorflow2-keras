```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                  'B': [5, np.nan, np.nan, 8],
                  'C': [10, 20, 30, 40]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
thresh = 0.25 * len(df)
df.dropna(axis=1, thresh=thresh)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'] = df['A'].fillna(value='0')
df
```

```python
df['B'] = df['B'].fillna(value=df['B'].mean())
df
```
